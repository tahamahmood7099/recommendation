```python
import pandas as pd
import numpy as np

from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python

df = pd.read_csv("/content/Dataset .csv")


df.head()
```

```python
print("Dataset Shape:", df.shape)

print("\nColumns:\n")
print(df.columns)

print("\nDataset Information:\n")
df.info()
```

```python
df.isnull().sum()
```

```python
for col in df.columns:

    if df[col].dtype == "object":

        df[col] = df[col].fillna(df[col].mode()[0])

    else:

        df[col] = df[col].fillna(df[col].median())

print("Missing values handled successfully.")
```

```python
data = df[[
    "Restaurant Name",
    "City",
    "Cuisines",
    "Average Cost for two",
    "Price range",
    "Aggregate rating"
]]

data.head()
```

```python
data["Average Cost for two"] = data["Average Cost for two"].astype(str)

data["Price range"] = data["Price range"].astype(str)
```

```python
data["Features"] = (

    data["City"] + " " +

    data["Cuisines"] + " " +

    data["Price range"]

)

data.head()
```

```python
cv = CountVectorizer()

feature_matrix = cv.fit_transform(data["Features"])

print(feature_matrix.shape)
```

```python
similarity = cosine_similarity(feature_matrix)

print(similarity.shape)
```

```python
def recommend_restaurant(name):

    if name not in data["Restaurant Name"].values:

        print("Restaurant not found.")

        return

    index = data[data["Restaurant Name"] == name].index[0]

    distances = list(enumerate(similarity[index]))

    distances = sorted(distances, key=lambda x: x[1], reverse=True)

    print("\nRecommended Restaurants\n")

    count = 0

    for i in distances[1:]:

        restaurant = data.iloc[i[0]]

        print("--------------------------------")

        print("Restaurant :", restaurant["Restaurant Name"])

        print("City :", restaurant["City"])

        print("Cuisine :", restaurant["Cuisines"])

        print("Average Cost :", restaurant["Average Cost for two"])

        print("Rating :", restaurant["Aggregate rating"])

        count += 1

        if count == 5:

            break
```

```python
data["Restaurant Name"].head(20)
```

```python
recommend_restaurant("Domino's Pizza")
```

```python
recommend_restaurant("Cafe Coffee Day")
```
